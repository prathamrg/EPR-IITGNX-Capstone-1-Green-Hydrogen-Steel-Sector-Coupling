```python
input_file = r"D:\Documents\Masters\IITGNX - Energy Policy and Regulation\Capstone Project\Final Simulation\Data\df_re_final.xlsx"
df_input = pd.read_excel(input_file, engine="openpyxl")

features = [
            'CoV hourly solar CF',
            'CoV hourly wind CF',
            'median hourly solar CF',
            'median hourly wind CF'
           ]

df_input_features = df_input[features]
df_input_features
```
